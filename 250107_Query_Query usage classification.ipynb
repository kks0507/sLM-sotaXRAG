```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.llms import HuggingFacePipeline
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
```

```python
# 모델 설정
model_id = "Bllossom/llama-3.2-Korean-Bllossom-3B"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
model.eval()
```

```python
# HuggingFacePipeline으로 LangChain에 통합
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=256,
    temperature=0.3,
    top_p=0.9,
    repetition_penalty=1.1
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.schema import SystemMessage, HumanMessage

# 프롬프트 설정 (LangChain)
classification_prompt = ChatPromptTemplate.from_messages([
    SystemMessage(content="너는 입력된 질문을 다음 중 하나로 분류해야 해: "
                          "'1' - 도서관 사용 질문, '2' - 학술적 지식 질문, '대답할 수 없습니다' - 비윤리적 질문.\n"
                          "예시:\n"
                          "'질문: 도서 대출 방법은?' -> '1'\n"
                          "'질문: 조선시대 농사법은?' -> '2'\n"
                          "'질문: 살인을 어떻게 저지르나요?' -> '대답할 수 없습니다'"),
    HumanMessage(content="질문: {question}")
])

# RunnableSequence로 LLM 연결
chain = classification_prompt | llm

# 입력 질문
question = "조선시대 농사법에 대해 알려주세요."
response = chain.invoke({"question": question})

print("분류 결과:", response.strip())
```
