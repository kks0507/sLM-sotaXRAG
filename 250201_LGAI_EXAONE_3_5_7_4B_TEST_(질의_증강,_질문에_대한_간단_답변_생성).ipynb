```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "LGAI-EXAONE/EXAONE-3.5-7.8B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 여기서 우리는 "System prompt"에 강력한 규칙을 기록.
system_prompt = """\
You are EXAONE model from LG AI Research, a helpful assistant.

[System Rules]
1) Always use 반말 (casual Korean). Do not use 존댓말 or formal endings.
2) Limit your search query to **four sentences or fewer**. If you exceed four sentences, stop output.
3) Write in the form of a "query" (ex: "뭐뭐 어떻게 찾으면 돼?")
4) Output must be placed strictly between the following delimiters:
   - ### Generate New Query
   - ### End Generate New Query
5) Do not include any other roles, explanations, or text outside those delimiters.
"""

user_prompt = "시간복잡도가 애플리케이션의 성능에 미치는 영향이 궁금해. 더 자세히 파고들 수 있는 검색 쿼리를 만들어 줘."

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# apply_chat_template를 통해 Chat 형태의 메시지를 model input으로 변환
input_ids = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt"
)

output = model.generate(
    input_ids.to("cuda"),
    eos_token_id=tokenizer.eos_token_id,
    max_new_tokens=128,
    do_sample=False,
)

decoded = tokenizer.decode(output[0])
print(decoded)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# 1. 모델 ID 설정
base_model_id = 'LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct'

# 2. BitsAndBytesConfig 설정
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True, # 4bit quantization으로 모델을 불러와 메모리 소모를 줄임
    bnb_4bit_use_double_quant=False, # True로 설정한다면, 이중 양자화를 통해 메모리를 극도로 절약할 수 있음
    bnb_4bit_quant_type="nf4", # 기존 fp4 보다 고급 형식의 normal float 4-bit
    bnb_4bit_compute_dtype=torch.bfloat16 # 계산에 사용할 데이터 타입
)

# 3. 모델 로드
model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config, torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto")
```

```python
# 4. 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    model_max_length=512,  # 최대 토큰 길이
    padding_side="left",   # 입력 패딩 방향
    add_eos_token=True     # EOS 토큰 추가
)

tokenizer.pad_token = tokenizer.eos_token
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
# LLM 파이프라인 설정 (LangChain과 통합)
llm_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=500,
    temperature=0.5,
    top_p=0.9,
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```

```python
system_prompt1 = """\
You are EXAONE model, a helpful assistant.

---As a scholar, I am looking for information on the latest scholarly research trends.
Your role as an "Academic Research Information Services Specialist" is to "generate new query" that enhance the scholarly expertise in my questions and help me find scholarly resources.
My question is "{question}".

---Then follow the instructions.
[System Rules]
<Rules>
After understanding the intent of my question, please "generate new query" in detail to make it professional and complete.
Always create roles, situations, and tasks within "generate new query", if possible.
</Rules>

<Strong Rules>
Output must be placed strictly between these delimiters:
   - ### Generate New Query
   - ### End Generate New Query
Do not include any other roles, explanations, or text outside these delimiters.
"Write in the form of a "query" (ex: "시간복잡도 분석이 앱 성능 최적화에 어떻게 기여하는지 설명해 주실 수 있나요?, 특히, 사용자 경험 측면에서 어떤 영향을 미치는지 구체적인 사례를 들어주세요.")
"generate new query" should be in half-speech form of "Korean(반말)".
"generate new query" should be "limited to two sentences".
</Strong Rules>
---
#[End]

"""
```

```python
from langchain_core.prompts import PromptTemplate
# PromptTemplate 설정
prompt1 = PromptTemplate(
    template=system_prompt1,
    input_variables=["question"]
)
```

```python
# 테스트 입력
query = "LLM과 Transformer의 차이를 알고 싶어."
formatted_prompt = prompt1.format(question=query)
response = llm(formatted_prompt)
print("LLM Response:\n", response)
```
