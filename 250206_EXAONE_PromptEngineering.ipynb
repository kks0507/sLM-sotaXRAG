## LLM 모델 준비

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# 1. 모델 ID 설정
base_model_id = 'LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct'

# 2. BitsAndBytesConfig 설정
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True, # 4bit quantization으로 모델을 불러와 메모리 소모를 줄임
    bnb_4bit_use_double_quant=False, # True로 설정한다면, 이중 양자화를 통해 메모리를 극도로 절약할 수 있음
    bnb_4bit_quant_type="nf4", # 기존 fp4 보다 고급 형식의 normal float 4-bit
    bnb_4bit_compute_dtype=torch.bfloat16 # 계산에 사용할 데이터 타입
)

# 3. 모델 로드
model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config, torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto")
```

```python
# 4. 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    model_max_length=512,  # 최대 토큰 길이
    padding_side="left",   # 입력 패딩 방향
    add_eos_token=True     # EOS 토큰 추가
)

tokenizer.pad_token = tokenizer.eos_token
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
# LLM 파이프라인 설정 (LangChain과 통합)
llm_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=500
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```

## RAG 체인 구축하기
Langchain 기반 RAG 체인을 구축합니다.

```python
PROMPT_TEMPLATE ="""\
You are a "Scholarly & Lecture Recommendation Assistant" specialized in analyzing academic resources and providing structured learning guidance.

--- Your task is to analyze the provided information (context) and user question to generate precise recommendations. My question is "{question}".
The relevant information is "{context}".

--- Then follow the instructions.
[System Rules]
<Rules>
After analyzing the intent of my question and the provided information, generate a structured recommendation response.
Always include **three key components** in your response:
1️⃣ **도서 추천**: 질문에 적합한 도서를 추천하고 그 이유를 설명하라.
2️⃣ **강의 추천**: 질문에 적절한 강의를 추천하고 그 이유를 설명하라.

Use a **fixed format** strictly without additional explanations.
</Rules>

<Strong Rules>
Output must be placed strictly between these delimiters:
   - ### Recommended Resources - Start
   - ### Recommended Resources - End
Do not include any other roles, explanations, or text outside these delimiters.
Follow the structured response format exactly as below:

### Recommended Resources - Start
1. 추천 도서 목록:
- 추천 1: [도서명] (저자)
  이유: [추천 이유]
- 추천 2: [도서명] (저자)
  이유: [추천 이유]
- 추천 3: [도서명] (저자)
  이유: [추천 이유]
- 추천 4: [도서명] (저자)
  이유: [추천 이유]

2. 추천 강의 목록:
- 추천 1: [강의명]
  이유: [추천 이유]
- 추천 2: [강의명]
  이유: [추천 이유]
- 추천 3: [강의명]
  이유: [추천 이유]
- 추천 4: [강의명]
  이유: [추천 이유]
### Recommended Resources - End

</Strong Rules>
---
#[End]
"""
```

```python
from langchain_core.prompts import PromptTemplate
# PromptTemplate 설정
prompt = PromptTemplate(
    template=PROMPT_TEMPLATE,
    input_variables=["context", "question"]
)
```

```python
# 두 개의 컬렉션에서 데이터를 검색하는 함수
def search_collections(query, syllabus_store, bibliography_store, top_k=4):
    query_embedding = embedding_model.embed_query(query)

    # 실라버스 컬렉션에서 검색
    syllabus_results = syllabus_store.similarity_search_by_vector(query_embedding, k=top_k)
    syllabus_context = "\n".join(
        f"강의명: {res.metadata['title']}, 강의 번호: {res.metadata['course_id']}, 주관 학과: {res.metadata['department']}\n{res.page_content}"
        for res in syllabus_results
    )

    # 서지 컬렉션에서 검색
    bibliography_results = bibliography_store.similarity_search_by_vector(query_embedding, k=top_k)
    bibliography_context = "\n".join(
        f"도서명: {res.metadata['title']}, 저자: {res.metadata['author']}, 출판사: {res.metadata['publisher']}, ISBN: {res.metadata['isbn']}\n{res.page_content}"
        for res in bibliography_results
    )

    # 두 결과를 합침
    combined_context = f"실라버스 정보:\n{syllabus_context}\n\n서지 정보:\n{bibliography_context}"
    return combined_context
```

```python
# 사용자 질의 처리 함수
def handle_query(query, syllabus_store, bibliography_store):
    # 검색 결과 가져오기
    context = search_collections(query, syllabus_store, bibliography_store)

    # Prompt에 맞게 데이터를 전달
    formatted_prompt = prompt.format(context=context, question=query)

    #LLM을 통해 결과 생성
    response = llm(formatted_prompt)

    # Check if the response is a string and convert to a dictionary if necessary
    if isinstance(response, str):
        response = {"generated_text": response}

    return response
```

```python
# 사용자 질의 테스트
query = "거시경제와 미시경제의 차이를 알 수 있는 강의 및 도서를 추천해줘."
response = handle_query(query, syllabus_store, bibliography_store)

print("추천 결과:")
print(response["generated_text"]) # Access the 'generated_text' key directly
```

```python
system_prompt_bookreview = """\
You are a "Career Coach & Book Review Specialist" who provides **personalized book reviews** based on users' academic and professional interests.

--- Your task is to generate a structured and insightful book review based on the user's input.
The user's book of interest is: "{booktitle}".
The relevant information is "{context}".

--- Then follow the instructions.

[System Rules]
<Rules>
After analyzing the given book title, provide **a personalized book review** that includes:
1️⃣ **책의 핵심 내용 요약**
2️⃣ **해당 책이 경제학 및 실무적 관점에서 어떻게 활용될 수 있는지 설명**
3️⃣ **연관된 학문적 배경(수강 과목, 연구 분야 등)과 연결하여 독자 맞춤형 설명 제공**
4️⃣ **이 책이 어떤 학습 방향을 가진 독자에게 적합한지 추천**

Use a **fixed format** strictly without additional explanations.
</Rules>

<Strong Rules>
Output must be placed strictly between these delimiters:
   - ### Personalized Book Review - Start
   - ### Personalized Book Review - End

Do not include any other roles, explanations, or text outside these delimiters.
Follow the structured response format exactly as below:

### Personalized Book Review - Start

📖 **"{booktitle}" 독서 서평**

"{booktitle}"은(는) [책의 핵심 내용 요약].
특히, [해당 분야에서 활용되는 방식]을 다루며, [연관된 학문적 개념]과 연계하여 독자의 이해를 돕습니다.
[이 책을 읽으면 얻을 수 있는 주요 인사이트]를 제공하며, [특정 학습 배경]을 가진 독자들에게 유용할 것입니다.
따라서, [이 책이 적합한 독자층]에게 추천됩니다.

### Personalized Book Review - End

</Strong Rules>
---
#[End]
"""
```

```python
from langchain_core.prompts import PromptTemplate
# PromptTemplate 설정
prompt = PromptTemplate(
    template=system_prompt_bookreview,
    input_variables=["context", "booktitle"]
)
```

```python
# 두 개의 컬렉션에서 데이터를 검색하는 함수
def search_collections(query,bibliography_store, top_k=1):
    query_embedding = embedding_model.embed_query(query)

    # 서지 컬렉션에서 검색
    bibliography_results = bibliography_store.similarity_search_by_vector(query_embedding, k=top_k)
    bibliography_context = "\n".join(
        f"도서명: {res.metadata['title']}, 저자: {res.metadata['author']}, 출판사: {res.metadata['publisher']}, ISBN: {res.metadata['isbn']}\n{res.page_content}"
        for res in bibliography_results
    )
```

```python
# 사용자 질의 처리 함수
def handle_query(query, bibliography_store):
    # 검색 결과 가져오기
    context = search_collections(query, bibliography_store)
    booktitle = query # Added: Assigning the query to 'booktitle'

    # Prompt에 맞게 데이터를 전달
    formatted_prompt = prompt.format(context=context, booktitle=booktitle) # Changed: Using 'booktitle'

    #LLM을 통해 결과 생성
    response = llm(formatted_prompt)

    # Check if the response is a string and convert to a dictionary if necessary
    if isinstance(response, str):
        response = {"generated_text": response}

    return response
```

```python
# 사용자 질의 테스트
query = """ "거시경제학의 이해"** (저자: 정운찬, 김영식) """
response = handle_query(query, bibliography_store)

print("추천 결과:")
print(response["generated_text"]) # Access the 'generated_text' key directly
```
